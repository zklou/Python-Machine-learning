```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
plt.rcParams['axes.unicode_minus']=False
y = pd.DataFrame(penguins_size.body_mass_g)

penguins_size = pd.read_csv('/Users/zklou/4404/4404A1/penguins_size.csv')
penguins_size.columns = ['species','island','culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g','sex',]
y.columns = ['species']

species_num = {'Adelie': 1, 'Chinstrap': 2, 'Gentoo': 3,}
penguins_size['species'] = penguins_size['species'].map(species_num)

color_theme = np.array(['lightsalmon', 'purple', 'green','blue'])
plt.subplot(1,2,1)
plt.scatter(x=penguins_size.flipper_length_mm, y=penguins_size.body_mass_g, c=color_theme[penguins_size.species], s=50)

plt.title('Penguin size, Palmer Station LTER')
plt.suptitle('Fllipper length and body mass for Adelle, Chinstrap, and Gentoo Penguins')
plt.legend(['Adelle','Chinstrap','Gentoo'])
plt.show()


penguins_size.head()

```
